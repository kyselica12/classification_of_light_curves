```python
from collections import defaultdict

def load_SDLCD_data(PATH, PATH_database, labels, regexes):
    with open(PATH, 'r') as f:
        data = json.load(f)
    
    with open(PATH_database, 'r') as f:
        database = json.load(f)

    name_database = {}

    for k in database:
        object_id = database[k]['object-id']
        object_name = database[k].get('object-name', '')
        name_database[object_id] = object_name
    
    
    labeled_data = {l: [] for l in labels}

    for d in data:
        object_id = list(d.keys())[0]
        object_name = name_database.get(object_id, '')

        if object_name == '':
            continue
        
        array = np.array([d[object_id][char + str(number)] for char in "ab" for number in range(1, 9)])

        for i, l in enumerate(labels):
            if re.search(regexes[i], object_name) is not None:
                labeled_data[l].append(array)
                break
        
    return labeled_data
```

```python
labeled_data = load_SDLCD_data(PATH, PATH_database, labels, regexes)
```

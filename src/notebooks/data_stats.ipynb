### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import tqdm
import time
import os
import sys
from collections import defaultdict
import pandas as pd
```

```python
DATA_PATH = "c:/Users/danok/work/dizertacka/resources/Fall_2021_2"
```

```python
sys.path.append('../')
from preprocessing.load_multi_array import load_multi_array
```

```python
objects = {"cylinder": ("falcon_9", "atlas_5", "atlas_V", "delta_4", "h_2a"), 
           "boxwing": ("navstar", "giove", "beidou", "astra", "meteosat", "globstar", "intelsat"),
           "nonsymetric": ("iridium", "orbcomm")}
```

```python
data = defaultdict(lambda : defaultdict(lambda: defaultdict(tuple)))
data_frame_data = []

# for folder in glob.iglob(f"{DATA_PATH}/*"):
#     folder_name = os.path.split(folder)[1]
    
files = [p for p in glob.iglob(f"{DATA_PATH}/*multi_array.npy")]

for file in tqdm.tqdm_notebook(files, desc=f"Folder {DATA_PATH}"):
    object_name = os.path.split(file)[1]
    arr = load_multi_array(file)
    n = int(len(arr))
    stat = np.sum(arr != 0, axis=1) / 300
    p_50 = np.sum(stat >0.5)/n 
    p_min = np.min(stat)
    p_max = np.max(stat)
    p_mean = np.mean(stat)

    for key1 in objects:
        for key2 in objects[key1]:
            if key2 in object_name.lower():
                data[key1][key2][object_name] = (n, p_max, p_min, p_mean, p_50) 
                data_frame_data.append([key1, key2, object_name, n, p_max, p_min, p_mean ,p_50])


df = pd.DataFrame(data_frame_data, columns=["type", "object_name", "name", "n_curves", "p_max", "p_min", "p_mean", "p_50"])
df.to_csv("stats_Fall_2021_2.csv", index=False)
```

```python
import sys
sys.path.append("../")

from config import load_config

import numpy as np
import matplotlib.pyplot as plt
from config import FilterConfig
from data.data_load import load_data
from data.filters import filter_data
from data.plot_light_curve import plot_curves
```

```python
data_cfg, _, filter_cfg = load_config()
data_cfg
```

```python
data_cfg.labels = ["cz-3", "cz-5", "titan", "falcon_9", "delta_4", "atlas_5"]
```

```python
data = load_data(data_cfg)
```

```python
filter_cfg = FilterConfig(
    n_bins=30,
    n_gaps=2, 
    gap_size=0, 
    rms_ratio= 0.1,
    non_zero_ratio=0.2)
filtered_data = filter_data(data, filter_cfg)
```

```python
for x in filtered_data:
    text = f"{x}:\n\ttotal: {len(data[x])}\n\tafter filter: {len(filtered_data[x])}\n\tpercentage: {round(len(filtered_data[x])/len(data[x]) * 100, 2)} %"
    print(text)
print("-------------------")
```

```python
for name in filtered_data:
    text = f"{name}:\n\ttotal: {len(data[name])}\n\tafter filter: {len(filtered_data[name])}\n\tpercentage: {round(len(filtered_data[name])/len(data[name]) * 100, 2)} %"
    print(text)
    plot_curves(-filtered_data[name][:10], n_cols=5)
```

```python
normalized = {}
for name in filtered_data:
    normalized[name] = filtered_data[name].copy()
    for i in range(len(filtered_data[name])):
        d = filtered_data[name][i]
        dd = d[d != 0]
        mean = np.mean(dd)
        std = np.std(dd)
        max = np.max(dd)
        min = np.min(dd)
        ddd = d.copy()
        ddd[d != 0] = (ddd[d != 0] - mean) / (std)
        normalized[name][i] = ddd


```

```python
for name in normalized:
    text = f"{name}:\n\ttotal: {len(data[name])}\n\tafter filter: {len(normalized[name])}\n\tpercentage: {round(len(normalized[name])/len(data[name]) * 100, 2)} %"
    print(text)
    plot_curves(-normalized[name][:10], n_cols=5, fit=True)
```
